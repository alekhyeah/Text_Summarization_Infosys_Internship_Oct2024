```python
from google.colab import userdata
import os
os.environ['GOOGLE_API_KEY'] = userdata.get('GOOGLE_API_KEY')
```

```python
from langchain.document_loaders import PyPDFLoader
from langchain.chains.summarize import load_summarize_chain
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(
      model="gemini-1.5-flash",
      temperature=0,
      max_tokens=None,
      timeout=None,
      max_retries=2
)
```

```python
def summarize_pdf(algorithm,pdf_file_path):
    loader = PyPDFLoader(pdf_file_path)
    docs = loader.load_and_split()
    chain = load_summarize_chain(llm, chain_type=algorithm)
    summary = chain.invoke(docs)

    return summary['output_text']
```

```python
algorithm=str(input("Enter the algorithm name:(refine or map_reduce) "))
path=str(input("Enter the path of the input text: "))
summary = summarize_pdf(algorithm,path)
summary
```
